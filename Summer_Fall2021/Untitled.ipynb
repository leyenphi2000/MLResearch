```python
#Libraries for importing and process data
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

#Library for visualization
import seaborn as sns 

avg = pd.read_csv('avg.csv', index_col=0)
avgRand = pd.read_csv('avgRand.csv', index_col=0)
```
