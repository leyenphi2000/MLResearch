```python
import glob
import os
import pandas as pd
import numpy as np

path = r'C:\Users\leyen\OneDrive\Desktop\School Works\Dr.Soibam Research\Summer2021\18states.bins'
os.chdir(path)
myFiles = glob.glob('*.txt')
```

```python
myFiles
```

```python
dat = {}

for c in range(0, len(myFiles)):
    data = pd.read_csv(path+'\\'+myFiles[c],delimiter = "\t", header = None)
    
    dat[c] = pd.DataFrame(index = data[0])

    numLines = []
    avgLen = []

    for i in range(0, len(data)):
        val = list(data[1].to_list()[i].split(","))
        val.append(0)
        line = []
        x = 1
        for n in range(0, len(val)-1):
            if val[n] == val[n+1]:
                x = x + 1
            else:
                line.append(x)
                x = 1
        numLines.append(len(line))
        avgLen.append(sum(line)/len(line))
    
    dat[c]["numLines"] = np.asarray(numLines)
    dat[c]["avgLen"] = np.asarray(avgLen)
```

```python
cell = []        
cells = []
for i in range(0, len(myFiles)):
    if i % 2 == 0:
        cell.append(myFiles[i].replace('.txt.boundary.bed.18core.states.bins.txt',''))
        cells.append(myFiles[i].replace('.txt.boundary.bed.18core.states.bins.txt',''))
    if i % 2 == 1:
        cells.append(myFiles[i].replace('.txt.boundary.bed.random.18core.states.bins.txt',''))
```

```python
norm = pd.DataFrame(columns = ["avgNum", "avgLen"],
                   index = cell)
for i in range(0, len(myFiles)):
    if i % 2 == 0:
        norm.loc[cells[i]] = list(dat[i].mean(axis='index'))
```

```python
rand = pd.DataFrame(columns = ["avgNum_rand", "avgLen_rand"],
                   index = cell)
for i in range(0, len(myFiles)):
    if i % 2 == 1:
        rand.loc[cells[i]] = list(dat[i].mean(axis='index'))
```
